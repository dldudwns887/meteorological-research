```python
# 데이터 패스 설정
path = '/home/papalio/test_research/python_edu/test_2024/test_2024/DATA/'

file_ta = 'sfc_grid_ta_202001010000.nc'

path_rn = '/home/papalio/test_research/python_edu/test_2024/test_2024/DATA/org/sgd/2020/01/01/'

file_rn = 'sfc_grid_rn_day_202001010000.nc'
```

```python
import xarray as xr

# 파일 경로 설정

# Dataset 열기
ds = xr.open_dataset(path+file_ta)

# 전체 Dataset 출력
print("=== Dataset 전체 구조 ===")
print(ds)

# 데이터 변수 및 속성 출력
print("\n=== 데이터 변수들 ===")
for var_name, var_data in ds.data_vars.items():
    print(f"변수 이름: {var_name}")
    print(var_data)

print("\n=== 속성 정보 ===")
for attr_name, attr_value in ds.attrs.items():
    print(f"{attr_name}: {attr_value}")

# 데이터 샘플 확인 (예: 'data' 변수)
if 'data' in ds:
    print("\n=== 'data' 변수 샘플 ===")
    print(ds['data'])

    # 데이터 통계 확인
    print("\n=== 'data' 변수 기본 통계 ===")
    print(f"최소값: {ds['data'].min().item()}")
    print(f"최대값: {ds['data'].max().item()}")
    print(f"평균값: {ds['data'].mean().item()}")
```
